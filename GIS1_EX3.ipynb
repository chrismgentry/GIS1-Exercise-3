```r
montco_tracts <- tracts("TN", county = "Montgomery")
```

```r
ggplot(montco_tracts) + geom_sf()
```

```r
population <- read.csv('https://raw.githubusercontent.com/chrismgentry/GIS1-Exercise-3/main/Data/mont_co_pop.csv', 
                        colClasses=c(Tract="character"))
head(population)
summary(population$under_5yo)
```

```r
tract_pop <- merge(x = montco_tracts, y = population, by.x = "NAME", by.y = "Tract", all = TRUE)
head(tract_pop)
```

```r
colors <- c('#440154', '#3a528b', '#20908d', '#5dc962', '#fde725')
ggplot(tract_pop, aes(fill = age5_18yo)) + geom_sf()
```

```r
schools <- read.csv('https://raw.githubusercontent.com/chrismgentry/GIS1-Exercise-3/main/Data/county_schools.csv')
head(schools)
summary(schools$ENROLLMENT)
```

```r
new_school <- data.frame(x  = -87.19495519624655, y = 36.56793328339189, name = "New School")
```

```r
ggplot() + 
  geom_sf(data = tract_pop, aes(fill = age5_18yo)) +
  scale_fill_viridis_c("K-12 Aged \nChildren", limits = c(0,2200), breaks = c(0,1100,2200), 
                       guide = guide_colourbar(title.hjust = 0.5)) +
  geom_point(data = schools, aes(x=X, y=Y), shape = 21, 
             size = 3, fill = "black", color = "white") +
  geom_point(data = new_school, aes(x=x, y=y,), shape = 8, size = 8, color = "red") +
  theme_minimal() +
  labs(title = "Number of K-12 Aged Children per Census Tract and School Populations") +
  north(tract_pop, location = "bottomleft", scale = 0.05, symbol = 12, anchor = c(x=-87.15, y=36.31)) + 
  scalebar(tract_pop, dist = 5, dist_unit = "mi", transform = TRUE, model = "WGS84", location = "bottomleft", st.dist = 0.02, st.size = 2, anchor = c(x= -87.65, y= 36.31)) +
  annotate("text", x = -87.22, y = 36.35, label = "Chris Gentry\n August 2021", size = 2) +
  theme(plot.title = element_text(hjust = 0.5))
```

```r
tn <- map_data('county', region = "tennessee")
states <- map_data('state')
```

```r
ggplot() + 
  geom_polygon(data = states, aes(x=long, y=lat, group=group), fill = "darkgray", color = "white") +
  geom_polygon(data = tn, aes(x=long, y=lat, group=group), fill = "lightgray", color = "white") +
  geom_sf(data = tract_pop, aes(fill = under_5yo)) +
  scale_fill_viridis_c("Children Aged 0-5", limits = c(0,1400), breaks = c(0,700,1400), 
                       guide = guide_colourbar(title.position = "top", title.hjust = 0.5, barwidth = 12)) + 
  geom_jitter(data = schools, aes(x=X, y=Y, size = ENROLLMENT, color = ENROLLMENT), shape = 16, width = 0.01, height = 0.01) +
  geom_point(data = new_school, aes(x=x, y=y,), shape = 8, size = 6, color = "red") +
  guides(color= guide_legend(title.position = "top", title.hjust = 0.5), size=guide_legend(title.position = "top", title.hjust = 0.5)) +
  scale_colour_viridis_c("Enrollment", option = "inferno", limits = c(0,2000), breaks = c(0,500,1000,1500,2000)) + 
  scale_size_continuous("Enrollment", limits = c(0,2000), breaks = c(0,500,1000,1500,2000)) +
  theme_void() + labs(title = "Number of Children Entering School through 2026 by Census Tract \nwith Current School Populations") +
  north(tract_pop, location = "bottomleft", scale = 0.05, symbol = 12, anchor = c(x=-87, y=36.645)) + 
  scalebar(tract_pop, dist = 5, dist_unit = "mi", transform = TRUE, model = "WGS84", location = "bottomleft", 
           st.dist = 0.03, st.size = 3, anchor = c(x= -87.65, y= 36.31)) +
  annotate("text", x = -87.64, y = 36.654, label = "Chris Gentry\n August 2021", size = 2) +
  coord_sf(xlim = c(-87.65,-86.99), ylim = c(36.30, 36.65)) +
  theme(legend.direction = "horizontal", legend.box = "vertical", legend.position = c(0.78, 0.15), legend.box.just = "center", 
        legend.spacing.x = unit(0.05, 'cm'), plot.title = element_text(hjust = 0.5), 
        plot.background = element_rect(fill = "darkgray", color = "white")) + 
  geom_text(data = new_school, aes(x=x,y=y, label = name), nudge_y = 0.025, nudge_x = 0.0, color = "white")
```
